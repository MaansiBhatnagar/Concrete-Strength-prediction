```python
# To apply logistic Regression model on my Dataset I need to add a feature in my dataset which classifies the cement as 'Strong' or 'Weak'. 
# So I will choose a threshold value and any value below that shall be weak and above that will be strong.
# So, the threshold value I am using is 0.5 for now. [Weak 711 Strong 319]
# using threshold value =0.4 , [Weak 515, Strong 515]
# Assuming concrete_df is your DataFrame
threshold = 0.4

# Create a new column 'Strength_Class' based on the threshold
concrete_df['Strength_Class'] = concrete_df['Strength'].apply(lambda x: 'Weak' if x <= threshold else 'Strong')

# Display the modified DataFrame
print(concrete_df) 
```

```python
# Evaluation of Logistic Regression model
import numpy as np
import sklearn.linear_model
from sklearn.metrics import precision_score, recall_score, f1_score

# concrete_df is a pandas DataFrame
concrete_df = concrete_df.sample(frac=1)  # Shuffle the DataFrame

# Split into training and validation sets
split_index = int(0.7 * len(concrete_df))
train_df = concrete_df[:split_index]
valid_df = concrete_df[split_index:]

# Check if the validation set is not empty
if not valid_df.empty:
    # Extract input features and labels for training set
    XTRAIN = train_df.iloc[:, :-1].values
    YTRAIN = train_df.iloc[:, -1].values

    # Extract input features and labels for validation set
    XVALID = valid_df.iloc[:, :-1].values
    YVALID = valid_df.iloc[:, -1].values

    # Create and fit a logistic regression model
    log_reg = sklearn.linear_model.LogisticRegression()
    log_reg.fit(XTRAIN, YTRAIN)

    # Predictions on training set
    YTRAIN_pred = log_reg.predict(XTRAIN)

    # Predictions on validation set
    if len(XVALID) > 0:  # Check if the validation set has samples
        YVALID_pred = log_reg.predict(XVALID)

        # Calculate metrics for training set
        precision_train = precision_score(YTRAIN, YTRAIN_pred)
        recall_train = recall_score(YTRAIN, YTRAIN_pred)
        f1_train = f1_score(YTRAIN, YTRAIN_pred)

        print("Precision on training set:", precision_train)
        print("Recall on training set:", recall_train)
        print("F1 score on training set:", f1_train)

        # Calculate metrics for validation set
        precision_valid = precision_score(YVALID, YVALID_pred)
        recall_valid = recall_score(YVALID, YVALID_pred)
        f1_valid = f1_score(YVALID, YVALID_pred)

        print("Precision on validation set:", precision_valid)
        print("Recall on validation set:", recall_valid)
        print("F1 score on validation set:", f1_valid)
    else:
        print("Validation set is empty.")
else:
    print("Training set is empty.")
```

```python
# Learning Curve for Logistic Regression model
import numpy as np
import sklearn.linear_model
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt

# concrete_df is a pandas DataFrame
concrete_df = concrete_df.sample(frac=1)  # Shuffle the DataFrame

# Split into training and validation sets
split_index = int(0.7 * len(concrete_df))
train_df = concrete_df[:split_index]
valid_df = concrete_df[split_index:]

# Check if the validation set is not empty
if not valid_df.empty:
    # Extract input features and labels for training set
    XTRAIN = train_df.iloc[:, :-1].values
    YTRAIN = train_df.iloc[:, -1].values

    # Extract input features and labels for validation set
    XVALID = valid_df.iloc[:, :-1].values
    YVALID = valid_df.iloc[:, -1].values

    # Create and fit a logistic regression model
    log_reg = sklearn.linear_model.LogisticRegression()
    log_reg.fit(XTRAIN, YTRAIN)

    # Predictions on training set
    YTRAIN_pred = log_reg.predict(XTRAIN)

    # Plot learning curve
    train_sizes, train_scores, valid_scores = learning_curve(log_reg, XTRAIN, YTRAIN, cv=5)

    # Calculate mean and standard deviation of training and validation scores
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    valid_scores_mean = np.mean(valid_scores, axis=1)
    valid_scores_std = np.std(valid_scores, axis=1)

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.title("Learning Curve")
    plt.xlabel("Training Examples")
    plt.ylabel("Score")

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1, color="r")
    plt.fill_between(train_sizes, valid_scores_mean - valid_scores_std,
                     valid_scores_mean + valid_scores_std, alpha=0.1, color="g")

    plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training Score")
    plt.plot(train_sizes, valid_scores_mean, 'o-', color="g", label="Validation Score")

    plt.legend(loc="best")
    plt.show()

    # Calculate metrics for training set
    precision_train = precision_score(YTRAIN, YTRAIN_pred)
    recall_train = recall_score(YTRAIN, YTRAIN_pred)
    f1_train = f1_score(YTRAIN, YTRAIN_pred)

    print("Precision on training set:", precision_train)
    print("Recall on training set:", recall_train)
    print("F1 score on training set:", f1_train)

    # Calculate metrics for validation set
    precision_valid = precision_score(YVALID, YVALID_pred)
    recall_valid = recall_score(YVALID, YVALID_pred)
    f1_valid = f1_score(YVALID, YVALID_pred)

    print("Precision on validation set:", precision_valid)
    print("Recall on validation set:", recall_valid)
    print("F1 score on validation set:", f1_valid)
else:
    print("Training set is empty.")
```

```python
# Building NN Model(16-8-4-1) and its evaluation
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import precision_score, recall_score, f1_score
import numpy as np
import matplotlib.pyplot as plt

# concrete_df is a pandas DataFrame
concrete_df = concrete_df.sample(frac=1)  # Shuffle the DataFrame

# Split into training and validation sets
split_index = int(0.7 * len(concrete_df))
train_df = concrete_df[:split_index]
valid_df = concrete_df[split_index:]

# Extract input features and labels for training set
XTRAIN = train_df.iloc[:, :-1].values
YTRAIN = train_df.iloc[:, -1].values

# Extract input features and labels for validation set
XVALID = valid_df.iloc[:, :-1].values
YVALID = valid_df.iloc[:, -1].values

# Build and compile your model
model = Sequential()
model.add(Dense(16, input_dim=XTRAIN.shape[1], activation='sigmoid'))
model.add(Dense(8, activation='relu'))  # First hidden layer with 4 neurons
model.add(Dense(4, activation='relu'))  # Second hidden layer with 3 neurons
model.add(Dense(1, activation='sigmoid'))  # Output layer with 1 neuron for binary classification
model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

# Learn the model from the training set
history = model.fit(XTRAIN, YTRAIN, epochs=256, verbose=1, validation_data=(XVALID, YVALID))

# Evaluate on the training set
accuracy_train = model.evaluate(XTRAIN, YTRAIN)
print("Accuracy on training set:", accuracy_train)

# Evaluate on the validation set
accuracy_valid = model.evaluate(XVALID, YVALID)
print("Accuracy on validation set:", accuracy_valid)

# Predict on the validation set
YVALID_pred = model.predict(XVALID)
YVALID_pred = (YVALID_pred > 0.5).astype(int)  # Convert probabilities to binary predictions

# Calculate precision, recall, and F1 score
precision = precision_score(YVALID, YVALID_pred)
recall = recall_score(YVALID, YVALID_pred)
f1 = f1_score(YVALID, YVALID_pred)

print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

# Plotting learning curve
plt.figure(figsize=(10, 6))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Learning Curve')
plt.legend()
plt.show()
```

```python
# Coding Custom Function to serve as a prediction model
# Creating Custom Function to make predictions
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score

def initialize_weights(layers):
    weights = [np.random.randn(layers[i], layers[i-1]) * np.sqrt(2/layers[i-1]) for i in range(1, len(layers))]
    biases = [np.zeros((layers[i], 1)) for i in range(1, len(layers))]
    return weights, biases

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def forward_propagation(inputs, weights, biases):
    layer_outputs = [inputs]
    
    for w, b in zip(weights[:-1], biases[:-1]):
        layer_outputs.append(relu(np.dot(w, layer_outputs[-1]) + b))
    
    output = sigmoid(np.dot(weights[-1], layer_outputs[-1]) + biases[-1])
    layer_outputs.append(output)
    
    return layer_outputs

def custom_model_predict(inputs, weights, biases):
    layer_outputs = forward_propagation(inputs.T, weights, biases)
    return layer_outputs[-1].ravel()

def train_model(X_train, Y_train, layer_dimensions, learning_rate=0.01, epochs=1000, batch_size=32):
    m = X_train.shape[0]
    weights, biases = initialize_weights(layer_dimensions)
    
    for epoch in range(epochs):
        # Shuffle training data for each epoch
        permutation = np.random.permutation(m)
        X_train_shuffled = X_train[permutation, :]
        Y_train_shuffled = Y_train[permutation]
        
        for i in range(0, m, batch_size):
            X_batch = X_train_shuffled[i:i+batch_size, :]
            Y_batch = Y_train_shuffled[i:i+batch_size]

            layer_outputs = forward_propagation(X_batch.T, weights, biases)

            # Backpropagation
            dA = layer_outputs[-1] - Y_batch.reshape(1, -1)
            dZ = dA * layer_outputs[-1] * (1 - layer_outputs[-1])
            dW = np.dot(dZ, layer_outputs[-2].T) / batch_size
            db = np.sum(dZ, axis=1, keepdims=True) / batch_size

            weights[-1] -= learning_rate * dW
            biases[-1] -= learning_rate * db

            for j in range(len(weights)-2, -1, -1):
                dZ = np.dot(weights[j+1].T, dZ) * (layer_outputs[j+1] > 0)
                dW = np.dot(dZ, layer_outputs[j].T) / batch_size
                db = np.sum(dZ, axis=1, keepdims=True) / batch_size

                weights[j] -= learning_rate * dW
                biases[j] -= learning_rate * db

    return weights, biases

# Example usage
X_TRAIN = train_df.iloc[:, :-1].values
Y_TRAIN = train_df.iloc[:, -1].values

# Define the architecture of the model
layer_dimensions = [X_TRAIN.shape[1], 16, 8, 4, 1]

# Train the model
trained_weights, trained_biases = train_model(X_TRAIN, Y_TRAIN, layer_dimensions, learning_rate=0.01, epochs=1000, batch_size=32)

# Predict on the validation set using the improved custom model
Y_VALID_pred_custom = custom_model_predict(XVALID, trained_weights, trained_biases)

# Convert probabilities to binary predictions
Y_VALID_pred_custom_binary = (Y_VALID_pred_custom > 0.5).astype(int)

# Calculate precision, recall, and F1 score
precision_custom = precision_score(YVALID, Y_VALID_pred_custom_binary)
recall_custom = recall_score(YVALID, Y_VALID_pred_custom_binary)
f1_custom = f1_score(YVALID, Y_VALID_pred_custom_binary)

print("Precision (Custom Model):", precision_custom)
print("Recall (Custom Model):", recall_custom)
print("F1 Score (Custom Model):", f1_custom)
```
